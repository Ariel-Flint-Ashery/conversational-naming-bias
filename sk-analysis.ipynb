```python
def option_probabilities(data,
                        option_list=options):
    
    prob_first_opt_all_sims = []
    
    for run_idx, run in data.items():
        if isinstance(run_idx, str):
            continue
        else:
            answer_list = run.get('tracker').get('answers')
            
            prob_first_opt = []
            prob_first_opt_acc = 0
            for t, pair in enumerate(answer_list):
                first_opt_count = pair.count(options[0])
                prob_first_opt_acc += first_opt_count
                prob_first_opt.append(prob_first_opt_acc/(2*(t+1)))
                
            prob_first_opt_all_sims.append(prob_first_opt)
                
    return prob_first_opt_all_sims
```

```python
homo_gay_evo = option_probabilities(data_homo)
```
